Fit a simple linear regression model (Salary ~ YearsExperience).
Report the estimated intercept and slope.
Predict the salary (in thousands) for someone with 4.5 years of experience.
Compute and interpret the R² value.
Note. Use both methods using scratch and scikit learn


```python
import pandas as pd

# Data dictionary
data = {
    "YearsExperience": [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7],
    "Salary (thousands)": [39.0, 46.0, 47.0, 52.0, 56.0, 64.0, 65.0, 67.0, 68.0, 70.0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV named q1.data
df.to_csv("q1data.cvs", index=False)

print("q1data created successfully!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df=pd.read_csv("q1data.csv")
df.head(3)
```

```python
df.columns
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.scatter(df["YearsExperience"],df["Salary (thousands)"])
plt.show()
```

Scratch

```python
x=df["YearsExperience"].values
y=df["Salary (thousands)"].values

xMean=np.mean(x)
yMean=np.mean(y)
```

```python
numerator=np.sum((x-xMean)*(y-yMean))
denominator=np.sum((x-xMean)**2)

slope=numerator/denominator

intercept=yMean-slope * xMean
```

```python
inputValue=4.5
predection=slope*inputValue + intercept
```

```python
yPrediction=slope*x+intercept

ssResidual =np.sum((y-yPrediction)**2)
ssTotal=np.sum((y-yMean)**2)

r2=1-(ssResidual/ssTotal)
```

```python
print("Manual Calculation Summary:")
print(f"Intercept: {intercept:.4f}")
print(f"Slope: {slope:.4f}")
print(f"Predicted Salary: {predection: .4f}")
print(f"R2 : {r2:.4f}")
```

Sk_Learn

```python
model=LinearRegression()
model.fit(df[["YearsExperience"]],df["Salary (thousands)"])
```

```python
predictedSalary = model.predict([[4.5]])
```

```python
r2=model.score(df[["YearsExperience"]],df["Salary (thousands)"])
```

```python
print("SK_Learn Summary:")
print(f"Intercept: {model.intercept_:.4f}")
print(f"Slope: {model.coef_[0]:.4f}")
print(f"Predicted Salary: {predictedSalary[0]}")
print(f"R2 : {r2:.4f}")
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.scatter(df["YearsExperience"],df["Salary (thousands)"])

ax.plot(df["YearsExperience"],model.predict(df["YearsExperience"].values.reshape(-1,1)),color="red")

plt.show()
```
