1.	Load an image from file and display it, convert to grayscale, resize it to specific size using OpenCV. 

```python
import cv2

image=cv2.imread("face.jpg")
greyedImage=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

cv2.imshow("Grey",greyedImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

2.	Create a blank image and draw basic shapes like rectangles and circles on it using OpenCV. 

```python
import cv2
import numpy as np

image=np.zeros((720,1280,3),dtype=np.uint8)

cv2.rectangle(image,(50,50),(200,230),(0,255,0),-1)
cv2.circle(image,(550,450),170,(140,40,40),-1)


cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

3.	Load an image and apply Gaussian blur, crop at specific region using OpenCV, NumPy array slicing. 

```python
import cv2
import numpy as np

image=cv2.imread("face.jpg")
image=cv2.GaussianBlur(image,(15,15),0)

image=image[100:800,70:750]

cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

4.	Load an image and add text to it using OpenCV, repeat this for creating a blank image and add text on that as well. 

```python
import cv2
import numpy as np

image=cv2.imread("face.jpg")
text="Abdullah Sonija 24K-0013"
font=cv2.FONT_HERSHEY_PLAIN
cv2.putText(image,text,(20,70),font,2.5,(0,0,0),2)

cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()

image2=np.zeros((720,1280,3),dtype=np.uint8)
text="Abdullah Sonija 24K-0013"
font=cv2.FONT_HERSHEY_PLAIN
cv2.putText(image2,text,(120,70),font,5,(10,210,90),3)

cv2.imshow("image2",image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

5.	Load a grayscale image and apply binary thresholding to it and the rotate at 60 degree using OpenCV. 

```python
import cv2

image=cv2.imread("face.jpg",cv2.IMREAD_GRAYSCALE)
ret,thresholdImage=cv2.threshold(image,100,180,cv2.THRESH_BINARY)

rot=cv2.getRotationMatrix2D((image.shape[1]/2,image.shape[0]/2),60,1)
thresholdImage=cv2.warpAffine(image,rot,(image.shape[1],image.shape[0]))

cv2.imshow("image",thresholdImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

6.	Load two images and blend them using OpenCV, and the convert it to grayscale and apply histogram equalization to enhance contrast. 

```python
import cv2

image1=cv2.imread("face.jpg")
image2=cv2.imread("pattern.jpg")

image1=cv2.resize(image1,(image2.shape[1],image2.shape[0]))

image=cv2.add(image1,image2)
image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
image=cv2.equalizeHist(image)

cv2.imshow("image +",image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
