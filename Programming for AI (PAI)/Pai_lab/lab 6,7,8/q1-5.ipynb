```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mobileDf = pd.read_csv("Mobile Reviews Sentiment.csv")
```

q1:

```python
avgRatingPerBrand = mobileDf.groupby("brand")["rating"].mean().sort_values()
brands = avgRatingPerBrand.index
ratings = avgRatingPerBrand.values

plt.figure(figsize=(10, 6))
plt.bar(brands, ratings, color='skyblue')
plt.xlabel("Brand")
plt.ylabel("Average Rating")
plt.title("Average Overall Rating per Mobile Brand")
plt.xticks(rotation=45)
plt.show()
```

q2:

```python
avgPricePerBrand = mobileDf.groupby("brand")["price_usd"].mean().sort_values()
brands = avgPricePerBrand.index
prices = avgPricePerBrand.values

plt.figure(figsize=(10, 6))
plt.bar(brands, prices, color='orange')
plt.xlabel("Brand")
plt.ylabel("Average Price (USD)")
plt.title("Average Price per Mobile Brand")
plt.xticks(rotation=45)
plt.show()
```

q3:

```python
plt.figure(figsize=(8, 6))
plt.hist(mobileDf["rating"], bins=5, color='green', edgecolor='black')
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Ratings")
plt.show()
```

q4:

```python
sentimentCounts = mobileDf["sentiment"].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(sentimentCounts, labels=sentimentCounts.index, autopct="%1.1f%%", colors=['red','gray','green'])
plt.title("Sentiment Distribution")
plt.show()
```

q5:

```python
meanPrice = mobileDf["price_usd"].mean()
meanRating = mobileDf["rating"].mean()

overpricedModels = mobileDf[(mobileDf["price_usd"] > meanPrice) & (mobileDf["rating"] < meanRating)]

plt.figure(figsize=(8, 6))
plt.scatter(mobileDf["price_usd"], mobileDf["rating"], alpha=0.5, label="All Models")
plt.scatter(overpricedModels["price_usd"], overpricedModels["rating"], color='red', marker='x', s=100, label="Overpriced Models")
plt.xlabel("Price (USD)")
plt.ylabel("Rating")
plt.title("Overpriced Models (High Price, Low Rating)")
plt.legend()
plt.show()
```

```python
ratingsBySentiment = [
    mobileDf[mobileDf['sentiment'] == 'positive']['rating'].dropna(),
    mobileDf[mobileDf['sentiment'] == 'neutral']['rating'].dropna(),
    mobileDf[mobileDf['sentiment'] == 'negative']['rating'].dropna()
]
labels = ['Positive', 'Neutral', 'Negative']

plt.figure(figsize=(10, 6))
plt.boxplot(ratingsBySentiment, labels=labels, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red'))
plt.title("Sentiment vs Rating Comparison (Box Plot)")
plt.xlabel("Sentiment")
plt.ylabel("Rating")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
