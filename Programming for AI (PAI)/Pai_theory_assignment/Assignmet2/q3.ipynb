```python
import pandas as pd
```

```python
data=pd.read_csv("Titanic-Dataset.csv")
```

```python
data.columns
```

```python
reportLines=[]
reportLines.append("Column, Dtype, % Missing ")
for col in data.columns:
    dtype=data[col].dtype
    missing=data[col].isna().mean()*100
    reportLines.append(f"{col}, {dtype}, {missing:.2f}% \n")
```

```python
with open("inspect_report.txt","w") as file:
    file.writelines(reportLines)
```

```python
print("Inspect report written sucessfully.")
print("preview (10 lines): ")
print("".join(reportLines[:10]))
```

```python
groupMedian=data.groupby(["Pclass","Sex"])["Age"].transform("median")
data["Age"]=data["Age"].fillna(groupMedian)
```

```python
data["Age"].info()
```

```python
if data["Embarked"].mode().size>0:
    embarkedMode=data["Embarked"].mode()[0]
else:
    embarkedMode="S"
data["Embarked"].fillna(embarkedMode)
```

```python
if "Cabin" in data.columns:
    data=data.drop(columns=["Cabin"])
```

```python
data.head(3)
```

```python
data["familySize"]=data["SibSp"]+data["Parch"]
data["IsAlone"]=(data["familySize"]==0).astype(int)
```

```python
data["Age"]=data["Age"].astype(int)
```

```python
data.head(2)
```

```python
data.to_csv("titanic_cleaned.csv",index=False)
print(f"Cleaned Data saved.")
```
