```python
import numpy as np
```

```python
data=[]
with open("sensor_data.csv","r") as file:
    for line in file:
        data.append(line.strip().split(","))
print(f"Loaded data: {len(data)} rows, {len(data[0])} columns")
```

```python
dataArr=np.array(data,dtype=float)
```

```python
dataArr[dataArr==-999]=np.nan
dataArr[(dataArr<0)|(dataArr>100)]=np.nan
    
```

```python
meanPerSensor=np.nanmean(dataArr,axis=0)
meanPerHour=np.nanmean(dataArr,axis=1)
```

```python
nanCount=np.isnan(dataArr).sum(axis=0)
wrostSensorIndex=np.argmax(nanCount)
wrostSensorInvalids=nanCount[wrostSensorIndex]
```

```python
print("Average moisture (per sensor):\n", meanPerSensor[:10], "...")  # first 10 to make the output simple
print("\nMedian moisture (first 10 hours):\n", meanPerHour[:10])
print(f"\nSensor with most invalid readings: Column {wrostSensorIndex} ({wrostSensorInvalids} NaNs)")
```

```python
minVal=np.nanmin(dataArr)
maxVal=np.nanmax(dataArr)
normalizedArr=(dataArr-minVal)/(maxVal-minVal)
```

```python
np.savetxt("sensor_data_normalized.csv",normalizedArr,delimiter=",",fmt="%.4f",header="Normalized Sensor Data",comments="")
```
