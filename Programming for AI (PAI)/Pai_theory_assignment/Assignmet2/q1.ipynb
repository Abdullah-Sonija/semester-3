```python
import requests as req
```

```python
with open("api.txt", "w") as file:
    file.write("https://api.cricapi.com/v1/currentMatches")

print("api.txt file created successfully!")
```

```python
try:
    with open("api.txt","r") as file:
        url=file.readline().strip() # using strip to leave the extra spaces
        print(url)
except FileNotFoundError:
    print("Error: 'api.txt' not found.")    
```

```python
param={
  "apikey": "ef61457d-fd8b-469f-a7a6-9cd089167c02",
  "offset": 0
}
```

```python
try:
    response=req.get(url,params=param)

    if response.status_code==200:
        data=response.json()
        print("API response recieved.")
        
        print("API Status Summary:")
        for key,value in data.items():
            if key not in ("data","apikey"):
                print(f"{key}: {value}")

        if data.get("status")!="success" or "data" not in data or not data["data"]:
            print("API Error.")
            print(f"Reason: {data.get("reason","unknown error")}")
        else:
            print("---Current Matches---")
            for match in data["data"]:
                print("-----------------------------------")
                print(f"Match Name  : {match.get("name","N/A")}")
                print(f"Status      : {match.get("status","N/A")}")
                print(f"Venue       : {match.get("venue","N/A")}")
                print(f"Date        : {match.get("date","N/A")}")
                print("-----------------------------------")

            scores=match.get("score")
            if scores and isinstance(scores,list):
                for innings in scores:
                    print(f"Inning      : {match.get("inning","N/A")}")
                    print(f"Runs        : {match.get("r","N/A")}")
                    print(f"Wickets     : {match.get("w","N/A")}")
                    print(f"Overs       : {match.get("o","N/A")}")
                    print("-----------------------------------")
            else:
                print("No score data avaiable yet.")
            print("--------------------------------------------------------------------------")
            
    else:
        print(f"HTTP Error: {response.status_code}")
        print(response.txt())

except req.exceptions.RequestException as e:
    print("Network Error: Unable to reach API.")
    print(f"Details: {e}")
```
